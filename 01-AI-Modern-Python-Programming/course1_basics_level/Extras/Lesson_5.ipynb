In Python, `is` and `in` are both operators, but they serve very different purposes:

### `is` Operator:
- **Purpose**: Tests for object identity.
- **Function**: Checks whether two variables point to the same object (i.e., have the same memory address).
- **Example**:




````python
a : int = 5
b : int = 6
print(id(a))
print(id(b))
a is not b

````

### `in` Operator:
- **Purpose**: Tests for membership.
- **Function**: Checks whether an element exists in a container (like a list, set, string, dictionary, etc.).
- **Example**:


````python
text = "hello"
print('h' in text)  # True ('h' is in the string "hello")
print('z' in text)  # False ('z' is not in the string "hello")
````

### Summary:
- **`is`**: Compares object identity.
- **`in`**: Checks for membership within a container.

#### PEMDAS

Alright, let's solve the equation step by step. We'll use the order of operations, often remembered by the acronym PEMDAS/BODMAS:

- **P**arentheses/**B**rackets
- **E**xponents (ie Powers and Square Roots, etc.)/**O**rders (ie Powers and Square Roots, etc.)
- **MD** Multiplication and Division (left-to-right)
- **AS** Addition and Subtraction (left-to-right)

Given the expression:

\[ 3 + 2 - 2 \times 4 \div 2 + 2 \]

Let's solve it step by step:

1. **Multiplication and Division**:
\[ 2 \times 4 = 8 \]
\[ 8 \div 2 = 4 \]

Replacing these values in the expression, we get:

\[ 3 + 2 - 4 + 2 \]

2. **Addition and Subtraction** (from left to right):
\[ 3 + 2 = 5 \]
\[ 5 - 4 = 1 \]
\[ 1 + 2 = 3 \]

So, the final answer is:

\[ 3 \]
